```python
animals = DataFrame(np.arange(16).reshape(4, 4),
                   columns=['W', 'X', 'Y', 'Z'],
                   index=['Dog', 'Cat', 'Bird', 'Mouse'])
```

```python
animals.loc[1:2, ['W', 'Y']] = np.nan 
```

```python
animals
```

```python
behavior_map = {'W': 'good', 'X': 'bad', 'Y': 'good','Z': 'bad'}
```

```python
animal_col = animals.groupby(behavior_map, axis=1)
animal_col.sum()
```

```python
behav_series = Series(behavior_map)
```

```python
behav_series
```

```python
animals.groupby(behav_series,axis=1).count()
```

Split apply combine
In essence, breaking up a dataframe, applying a function to it and then combining it again.


```python
dframe_wine
```

```python
def ranker(df):
    df['alc_rank']= np.arange(len(df))+1
    return df
```

```python
dframe_wine.sort_values('alcohol',ascending = False, inplace = True)
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
dframe_wine
```

Below is a method for creating data sets from a string.  That is why the instructor used the IO module.  It allows you to create a list as a comment string and save it as a variable, make a data frame out of it by using pandas' read_table in conjunction with StringIO to convert the string into a dataframe.

```python
data="""\Sample Animal Intelligence
1 Dog Smart
2 Dog Smart
3 Cat Dumb
4 Cat Dumb
5 Dog Dumb
6 Cat Smart"""
```

```python
dframe = pd.read_table(StringIO(data),sep='\s+')
```

```python
dframe
```

.crosstab creates a frequency table by the items you pass through separated by commas in dframe dot format shown below.  You could have easily used a pivot table to do this too.

```python
pd.crosstab(dframe.Animal,dframe.Intelligence, margins=True)
```
