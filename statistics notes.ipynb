Types of variables
Numerical - number values that can sometimes be manipulated mathematically.
Discrete - whole, non-negative that can only take numerical values with jumps.
Continuous - can take many values
Categorical - made from variables that are categories and their values are called levels.
Ordinal - is like yes, no, maybe variables also called binary in other fields.
Explanatory and response - a variable included in data collection to test an assumption or correlation.
Confounding or “lurking” variable - correlated with both the explanatory and response variables.
Example: The confounding variable of sunlight exposure would correlate to both sunscreen use and skin cancer rates.


Variable relationships

Dependent - another word for associated variables.

Independent - variables that are not associated

Data collection principles
A clearly laid out research question is helpful in identifying what cases should be studied and what variables are important.

Our question or problem should refer to a target population.  This is the subject of our research.  Since it is usually impractical to attempt to collect data on an entire population of anything, we use a sample to represent it.

When we come to conclusions based on only one or two known cases rather than a random sample of a population, it is said to be based on anecdotal evidence.  This data is rarely representative of the target population and only serves to help prove an individual’s conclusion right to him or herself.  It is important to note that anecdotal evidence may be correct and verifiable, but is still false when applied to the entire target population.

When sampling the target population, it is important to try to do so in as mathematically random a way as possible.  This is because a biased sample can skew our results.

A simple random sample is the most basic method and is equivalent to a raffle to select cases.
This means each member of the target population has an equal chance of being selected.

There is still bias that can pop up in simple random samples like non-response issues in surveys skewing data to characteristics of only those who respond or convenience sampling in which the sample is not representative of harder to reach population members.
A good example of these hidden biases are product reviews on amazon.  How much more likely is the average person willing to take the time to write a good review versus a bad one?  Let that sink in.

Observational studies and experiments

The two types of data collection are observation and experiment.

Observational studies do not directly interfere with how the data arise.
It provides evidence of naturally occurring associations, but cannot show causal connections.

For causal connections, we require experiments.  There will usually be explanatory and response variables included.

A randomized experiment = sample cases randomly assigned to control or test group of an experiment in which an explanatory and response variable are collected.
Please remember that correlation is NOT causation and causation can only be inferred from a randomized experiment.

Don’t make causal conclusions on observational studies without conducting an experiment.  Example is double swap verify principle.
Observational studies are for discovering associations in order to experiment on those associations further.


Observational studies are either prospective or retrospective.
Prospective study occurs as events unfold.
Retrospective study occurs after the events.

Sampling methods

Simple, stratified, and cluster sampling are the main three.
Simple is the most intuitive.
Each case in the population has an equal chance of being included and knowing that a case is included in a sample does not provide useful information about which other cases are included.

Stratified sampling is divide and conquer based.  The population is grouped into strata (similar cases grouped together).  Then a second sampling method is used on each strata (usually simple).  This is useful when cases in each stratum are similar with respect to the outcome of interest.  It is , however, more difficult to analyze.

Cluster sampling groups observations into clusters and randomly samples some of the clusters.
This is most used when there is a lot of variability between cases in each cluster but all clusters look similar. Analyzing is also more difficult.

Multistage sampling would be using a simple random sampling inside of a cluster sample.  This makes the cluster sample more like a stratified sample without the groupings by similarities.

Experimental studies that include randomization in regards to which cases are assigned to test or control group are called a randomized experiment.  These are good tools for determining causal connections between two variables and they are built on four principles.

Controlling - Minimize variables.  

Randomization - Randomize which cases are in control and treatment groups.

Replication - Repeat the experiment, use a large sample size, or have others repeat the experiment.

Blocking - compensating for other variables that might impact the explanatory variable by separating the sample group into the other potentially salient variable and then randomizing which cases are assigned to control and treatment groups within each block.

To help reduce bias in studies, researchers don’t let the patient know if they are in the control group or the treatment group and everyone gets a treatment.  Those in the control group get the placebo.  This is a blind study.  When the doctors and researchers are also prevented from knowing, it is called double blind.

Scatter plots for paired data - tuples?



Scatterplots provide a case-by-case view of data for two numerical variables.  Each point will represent a single case and each variable will be one of the two axis.  

A dot plot is a one variable scatterplot.  The mean avg is usually used on the distribution of the dot plot.  This is often labeled as x with a line over it and called the sample mean.
If there are a lot of cases, a stacked dot plot maybe used to help diffrientiate between the cases.


The mean of the entire population is represented by u (greek letter mu) and is calculated the same way. To distinguish between various variables the mean may us a subscirpt of x to identify each.

A simple mean turns into a weighted mean if the avg is done without nesting to reach the unweighted value of the individual object in the cases.  As an example the per capita income of people across all counties in the US must be averaged first by county then the averages averaged again for the total to avoid a weighted mean.

A histogram looks like a bar chart but is more closely related to a scatter plot or dot plot.  It takes the dots and dumps them into categories or bins that contain a range.  Anything that falls on a boundary dumps to the lower range bin.  Histograms are usful for showing data density and shape of the data distribution.  Histograms are also good for finding mode.  Mode is a prominent peak in the distibution.  A histogram can be unimodal, bimodal, or multimodal.
